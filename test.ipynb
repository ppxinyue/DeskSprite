```python
from PIL import Image
import imageio

img = Image.open("cat.png")

frames = []

# 简单缩放动画
for scale in [1, 1.05, 1.1, 1.05, 1]:
    w, h = img.size
    resized = img.resize((int(w*scale), int(h*scale)))
    frames.append(resized)

imageio.mimsave("cat_zoom.gif", frames, duration=0.1)
```
